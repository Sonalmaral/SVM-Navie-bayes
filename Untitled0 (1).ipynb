Q1. What is a Support Vector Machine (SVM)?
>> SVM is a supervised learning algorithm used for classification and regression.
It works by finding the optimal hyperplane that best separates data points of different classes in a high-dimensional space.
The goal: maximize the margin — the distance between the hyperplane and the nearest data points (support vectors).

Q2. Difference between Hard Margin and Soft Margin SVM
>> Type	Description	Use Case
a. Hard Margin	Assumes data is perfectly linearly separable; no misclassification allowed.
Ideal for clean, noise-free datasets.

b. Soft Margin	Allows some misclassifications by introducing a penalty term C to handle noise/outliers.	Works better with real-world, imperfect data.

Q3. Mathematical Intuition Behind SVM
>> Minimize 1/2 ||w|| square
where 2/||w|| is maximize margin
||w|| square  is minimise margine

Q4. Role of Lagrange Multipliers in SVM
>> Lagrange multipliers convert the constrained optimization problem into an unconstrained one (dual form).
They help find the optimal w and b by assigning weights (α_i) to each constraint.
Only points with α_i > 0 become support vectors.

Q5. What are Support Vectors in SVM?
>> Support vectors are the data points closest to the hyperplane.
They determine the position and orientation of the decision boundary.
Removing non-support vectors doesn’t affect the boundary.

Q6. What is a Support Vector Classifier (SVC)?
>> SVC is the classification implementation of SVM.
It classifies data by finding the optimal hyperplane separating different classes — can use linear or non-linear kernels.

Q7. What is a Support Vector Regressor (SVR)?
>> SVR is the regression version of SVM.
It fits the data within a tube (margin of tolerance) and ignores small errors.
Only points outside the margin affect the model.

Q8. What is the Kernel Trick in SVM?
>> The kernel trick maps input data into a higher-dimensional space without explicitly computing the transformation.
This allows SVM to handle non-linear separable data using functions like RBF or polynomial kernels.

Q9. Compare Linear, Polynomial, and RBF Kernels
>> a. Linear: Fast, simple and use for linearly separable data
b. Polynomial: Captures interactions between features and use for Moderate non-linearity
c. RBF (Gaussian): when when x_i - x_j

Q10. Effect of the C Parameter in SVM
>> C controls the trade-off between maximizing margin and minimizing classification error.

High C → small margin, tries to classify all points correctly (overfitting).

Low C → larger margin, allows misclassification (underfitting).

Q11. Role of Gamma Parameter in RBF Kernel SVM
>> Gamma defines how far the influence of a single training example reaches.

High gamma → narrow influence, complex boundary (overfitting).

Low gamma → broad influence, smoother boundary (underfitting).

Q12. What is the Naïve Bayes classifier, and why is it called "Naïve"?
>> It’s a probabilistic classifier based on Bayes’ Theorem that assumes all features are independent given the class label.
It’s called “naïve” because in reality, features are rarely completely independent — but the assumption simplifies computation.

Q13. What is Bayes’ Theorem?
>>P(A∣B)=P(B∣A)⋅P(A)/P(B)


Q14. Differences — Gaussian, Multinomial, and Bernoulli Naïve Bayes
>>
a. Type : Gaussian: data type is Continuous , eg. Sensor data, numeric features

b. type: Multinomial: data type: Discrete counts, eg. Word counts (text data)

c. type: Bernoulli:  data type: Binary, eg. Binary features (word presence/absence)

Q15. When to use Gaussian Naïve Bayes
>> Use Gaussian NB when features are continuous and approximately normally distributed, e.g. height, temperature, age.

Q16. Key assumptions made by Naïve Bayes
>> Feature independence: All features contribute independently to the outcome.

Equal importance: Each feature contributes equally to prediction.

Q17. Advantages and Disadvantages
>> Advantages

Simple and fast

Works well with small data

Effective for high-dimensional text data

Requires little training data

Disadvantages

Strong independence assumption rarely holds

Poor performance with correlated features

Q18: Why is Naïve Bayes good for text classification?
>> Text features (word frequencies) are mostly independent.

Fast and efficient on high-dimensional data.

Works well with sparse data like TF-IDF or bag-of-words.

Q19. How does Laplace Smoothing help in Naïve Bayes?
>> Laplace smoothing prevents zero probability for unseen feature-class combinations.
It adds a small constant (usually 1) to frequency counts:

````python
#Q1. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy:
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# You can inspect the data if needed
# print('Features:', iris.feature_names)
# print('Target Names:', iris.target_names)
# print('First 5 samples of X:\n', X[:5])
# print('First 5 samples of y:\n', y[:5])

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the SVM classifier
# Using a linear kernel for simplicity, but 'rbf' (Gaussian) is also very common for non-linear separation
svm_classifier = SVC(kernel='linear', random_state=42)

# Train the classifier on the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = svm_classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM Classifier on Iris dataset: {accuracy:.4f}")
````

````python
#Q2.Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, thencompare their accuracies
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Wine dataset
wine = datasets.load_wine()
X = wine.data
y = wine.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Train SVM with Linear Kernel ---
linear_svm = SVC(kernel='linear', random_state=42)
linear_svm.fit(X_train, y_train)
y_pred_linear = linear_svm.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print(f"Accuracy of SVM (Linear Kernel) on Wine dataset: {accuracy_linear:.4f}")

# --- Train SVM with RBF Kernel ---
rbf_svm = SVC(kernel='rbf', random_state=42)
rbf_svm.fit(X_train, y_train)
y_pred_rbf = rbf_svm.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)
print(f"Accuracy of SVM (RBF Kernel) on Wine dataset: {accuracy_rbf:.4f}")

# Compare accuracies
if accuracy_linear > accuracy_rbf:
    print("\nLinear Kernel performed better on this dataset.")
elif accuracy_rbf > accuracy_linear:
    print("\nRBF Kernel performed better on this dataset.")
else:
    print("\nBoth kernels performed equally well on this dataset.")
````

````python

#Q3. Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using MeanSquared Error (MSE):import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# Load the California Housing dataset
housing = datasets.fetch_california_housing()
X = housing.data
y = housing.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features. SVR is sensitive to the scale of the data.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize the SVR model
# Using a radial basis function (RBF) kernel, which is common for regression.
# C and epsilon are important hyperparameters to tune for SVR.
svr_model = SVR(kernel='rbf', C=1.0, epsilon=0.2)

# Train the SVR model on the scaled training data
svr_model.fit(X_train_scaled, y_train)

# Make predictions on the scaled test set
y_pred = svr_model.predict(X_test_scaled)

# Evaluate the model using Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error of SVR on California Housing dataset: {mse:.4f}")
````

````python
# Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decisionboundary:import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
# 1. Generate a synthetic dataset (not linearly separable)
X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Train SVM with Polynomial Kernel
# degree: the degree of the polynomial kernel function (e.g., 2 for quadratic, 3 for cubic)
# C: penalty parameter C of the error term.
svm_poly = SVC(kernel='poly', degree=3, C=1.0, random_state=42)
svm_poly.fit(X_train, y_train)

# 3. Visualize the decision boundary
def plot_decision_boundary(X, y, model, title):
    # Create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Predict class for each point in the mesh
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdBu)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu, edgecolors='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.show()

plot_decision_boundary(X, y, svm_poly, 'SVM with Polynomial Kernel Decision Boundary')
````

````python
#Q6. Write a Python program to train a Multinomial Naïve Bayes classifier for text classification using the 20Newsgroups dataset.
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Load the 20 Newsgroups dataset
# We'll select a few categories for simplicity, or load all if not specified
categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']
newsgroups_train = fetch_20newsgroups(subset='train', categories=categories, shuffle=True, random_state=42)
newsgroups_test = fetch_20newsgroups(subset='test', categories=categories, shuffle=True, random_state=42)

X_train_docs = newsgroups_train.data
y_train = newsgroups_train.target
X_test_docs = newsgroups_test.data
y_test = newsgroups_test.target

print(f"Training data size: {len(X_train_docs)} documents")
print(f"Test data size: {len(X_test_docs)} documents")
print(f"Categories: {newsgroups_train.target_names}")

# Convert text documents to a matrix of token counts (feature vectors)
# Multinomial Naïve Bayes works well with count features
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(X_train_docs)
X_test = vectorizer.transform(X_test_docs)

print(f"Shape of training feature matrix: {X_train.shape}")
print(f"Shape of test feature matrix: {X_test.shape}")

# Initialize the Multinomial Naïve Bayes classifier
mnb_classifier = MultinomialNB()

# Train the classifier on the training data
mnb_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = mnb_classifier.predict(X_test)

# Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of Multinomial Naïve Bayes Classifier on 20 Newsgroups dataset: {accuracy:.4f}")

# Print a classification report for more detailed metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=newsgroups_test.target_names))
````

````python

#Q7. Write a Python program to train an SVM Classifier with different C values and compare the decisionboundaries visually
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# 1. Generate a synthetic dataset for classification
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=2,
                           random_state=42, n_clusters_per_class=1)

# Split data into training and testing sets (though we'll use all data for visualization for simplicity)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Define a range of C values to compare
C_values = [0.1, 1, 10, 100]

# 3. Helper function to plot decision boundaries
def plot_decision_boundary(X, y, model, ax, title):
    # Create a mesh to plot in
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Predict class for each point in the mesh
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    ax.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.RdBu)
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu, edgecolors='k')
    ax.set_title(title)
    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_xticks(())
    ax.set_yticks(())

# 4. Train SVMs with different C values and plot their decision boundaries
fig, axes = plt.subplots(1, len(C_values), figsize=(4 * len(C_values), 4))

for i, C in enumerate(C_values):
    svm_classifier = SVC(kernel='linear', C=C, random_state=42)
    svm_classifier.fit(X_train, y_train)

    # Predict on the full dataset for visualization purposes
    plot_decision_boundary(X, y, svm_classifier, axes[i], f'SVM (C={C})')

plt.suptitle('SVM Decision Boundaries with Different C Values', y=1.02, fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.98]) # Adjust layout to prevent title overlap
plt.show()
````

````python
#Q8.Write a Python program to train a Bernoulli Naïve Bayes classifier for binary classification on a dataset withbinary features=import numpy as np
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification

# 1. Generate a synthetic dataset with binary features
# make_classification can generate binary features by setting n_informative=2, n_redundant=0, n_clusters_per_class=1
# We then convert these features to binary (0 or 1) by thresholding.
X, y = make_classification(n_samples=200, n_features=10, n_informative=5, n_redundant=0,
                           n_clusters_per_class=1, random_state=42, n_classes=2)

# Convert features to binary (0 or 1) - crucial for Bernoulli Naive Bayes
# We'll use a simple threshold, e.g., features > 0 become 1, features <= 0 become 0
X_binary = (X > 0).astype(int)

print(f"Shape of binary features (X_binary): {X_binary.shape}")
print(f"First 5 samples of binary features:\n{X_binary[:5]}")
print(f"First 5 samples of target (y): {y[:5]}\n")

# 2. Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_binary, y, test_size=0.3, random_state=42)

# 3. Initialize the Bernoulli Naïve Bayes classifier
# binarize=None: Assumes input features are already binary. If features are not binary,
#                BernoulliNB can binarize them internally using a threshold.
#                Here, we've already binarized X_binary explicitly.
bnb_classifier = BernoulliNB(binarize=None)

# 4. Train the classifier on the training data
bnb_classifier.fit(X_train, y_train)

# 5. Make predictions on the test set
y_pred = bnb_classifier.predict(X_test)

# 6. Evaluate the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Bernoulli Naïve Bayes Classifier: {accuracy:.4f}")
````

````python
#Q9. Write a Python program to apply feature scaling before training an SVM model and compare results withunscaled data=
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("--- SVM with Unscaled Data ---")
# Initialize and train SVM on unscaled data
svm_unscaled = SVC(kernel='linear', random_state=42)
svm_unscaled.fit(X_train, y_train)

# Make predictions and evaluate accuracy
y_pred_unscaled = svm_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)
print(f"Accuracy of SVM (Unscaled Data): {accuracy_unscaled:.4f}")


print("\n--- SVM with Scaled Data ---")
# Apply feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train SVM on scaled data
svm_scaled = SVC(kernel='linear', random_state=42) # Using linear kernel for direct comparison
svm_scaled.fit(X_train_scaled, y_train)

# Make predictions and evaluate accuracy
y_pred_scaled = svm_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)
print(f"Accuracy of SVM (Scaled Data): {accuracy_scaled:.4f}")


print("\n--- Comparison ---")
if accuracy_scaled > accuracy_unscaled:
    print(f"Feature scaling improved accuracy by: {(accuracy_scaled - accuracy_unscaled):.4f}")
elif accuracy_unscaled > accuracy_scaled:
    print(f"Unscaled data performed better, but this is less common: {(accuracy_unscaled - accuracy_scaled):.4f}")
else:
    print("Both scaled and unscaled data achieved the same accuracy.")
````

````python
#Q30. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before andafter Laplace Smoothing=import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

print("Generating a synthetic dataset...")
# Generate a synthetic dataset with some features that might have low variance for certain classes
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=0,
                           n_clusters_per_class=1, random_state=42)

# Introduce some features with very low variance for a specific class to highlight the effect
# For example, make a feature almost constant for class 0
X[y == 0, 0] = np.random.normal(loc=0.1, scale=0.01, size=(X[y == 0].shape[0]))
X[y == 1, 0] = np.random.normal(loc=1.0, scale=1.0, size=(X[y == 1].shape[0]))

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")

print("\n--- Gaussian Naïve Bayes without var_smoothing (default) ---")
# Train GaussianNB with default var_smoothing (usually very small, effectively no smoothing)
gnb_no_smoothing = GaussianNB()
gnb_no_smoothing.fit(X_train, y_train)
y_pred_no_smoothing = gnb_no_smoothing.predict(X_test)

accuracy_no_smoothing = accuracy_score(y_test, y_pred_no_smoothing)
print(f"Accuracy without var_smoothing: {accuracy_no_smoothing:.4f}")
print("Classification Report without var_smoothing:")
print(classification_report(y_test, y_pred_no_smoothing))

print("\n--- Gaussian Naïve Bayes with var_smoothing (e.g., 1e-9) ---")
# Train GaussianNB with a chosen var_smoothing value
# A common practice is to use a small value like 1e-9 or similar.
gnb_with_smoothing = GaussianNB(var_smoothing=1e-9) # Adjust this value to see more pronounced effects
gnb_with_smoothing.fit(X_train, y_train)
y_pred_with_smoothing = gnb_with_smoothing.predict(X_test)

accuracy_with_smoothing = accuracy_score(y_test, y_pred_with_smoothing)
print(f"Accuracy with var_smoothing: {accuracy_with_smoothing:.4f}")
print("Classification Report with var_smoothing:")
print(classification_report(y_test, y_pred_with_smoothing))

print("\n--- Comparison of Accuracies ---")
if accuracy_with_smoothing > accuracy_no_smoothing:
    print(f"Using var_smoothing improved accuracy by: {(accuracy_with_smoothing - accuracy_no_smoothing):.4f}")
elif accuracy_no_smoothing > accuracy_with_smoothing:
    print(f"Default (less) var_smoothing performed better by: {(accuracy_no_smoothing - accuracy_with_smoothing):.4f}")
else:
    print("Both models achieved the same accuracy.")
````

````python
#Q.11 Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C,gamma, kernel)=import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# It's good practice to scale the data, especially for SVMs, before GridSearchCV
# as it helps the optimization process and prevents issues with some kernels (like RBF).
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the SVM classifier
svm = SVC(random_state=42)

# Define the hyperparameters and their possible values for tuning
# C: Regularization parameter. The strength of the regularization is inversely proportional to C.
# gamma: Kernel coefficient for 'rbf', 'poly' and 'sigmoid'. Affects how much a single training example influences the decision boundary.
# kernel: Specifies the kernel type to be used in the algorithm. 'linear', 'poly', 'rbf', 'sigmoid'.
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.001, 0.01, 0.1, 1],
    'kernel': ['rbf', 'linear'] # Often 'rbf' and 'linear' are good starting points
}

# Initialize GridSearchCV
# estimator: The model to tune (SVC)
# param_grid: The dictionary of hyperparameters to search
# cv: Number of folds for cross-validation
# verbose: Controls the verbosity: the higher, the more messages.
# n_jobs: Number of jobs to run in parallel (-1 means using all processors).
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)

print("Starting GridSearchCV for hyperparameter tuning...")
# Fit GridSearchCV to the scaled training data
grid_search.fit(X_train_scaled, y_train)

print("\nGridSearchCV complete.")

# Print the best parameters found
print(f"Best parameters found: {grid_search.best_params_}")

# Print the best cross-validation score
print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}")

# Get the best estimator (model) from GridSearchCV
best_svm_model = grid_search.best_estimator_

# Make predictions on the scaled test set using the best model
y_pred = best_svm_model.predict(X_test_scaled)

# Evaluate the best model's performance on the test set
accuracy = accuracy_score(y_test, y_pred)
print(f"\nTest set accuracy with best SVM model: {accuracy:.4f}")
print("\nClassification Report on Test Set:")
print(classification_report(y_test, y_pred))
````

````python
#Q16.Write a Python program to train an SVM Classifier using One-vs-Rest (OvR) and One-vs-One (OvO)strategies on the Wine dataset and compare their accuracy=
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the Wine dataset
wine = datasets.load_wine()
X = wine.data
y = wine.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature scaling is important for SVMs
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- SVM with One-vs-Rest (OvR) Strategy ---")
# For OvR, we use OneVsRestClassifier wrapping an SVC
# The base estimator (SVC) will be trained for each class against all others.
ovr_svm = OneVsRestClassifier(SVC(kernel='linear', random_state=42))
ovr_svm.fit(X_train_scaled, y_train)
y_pred_ovr = ovr_svm.predict(X_test_scaled)
accuracy_ovr = accuracy_score(y_test, y_pred_ovr)
print(f"Accuracy of SVM (One-vs-Rest): {accuracy_ovr:.4f}")

print("\n--- SVM with One-vs-One (OvO) Strategy ---")
# For OvO, sklearn's SVC handles this automatically for multiclass problems
# It trains n_classes * (n_classes - 1) / 2 classifiers.
ovo_svm = SVC(kernel='linear', decision_function_shape='ovo', random_state=42)
ovo_svm.fit(X_train_scaled, y_train)
y_pred_ovo = ovo_svm.predict(X_test_scaled)
accuracy_ovo = accuracy_score(y_test, y_pred_ovo)
print(f"Accuracy of SVM (One-vs-One): {accuracy_ovo:.4f}")

print("\n--- Comparison ---")
if accuracy_ovr > accuracy_ovo:
    print(f"One-vs-Rest (OvR) performed better by: {(accuracy_ovr - accuracy_ovo):.4f}")
elif accuracy_ovo > accuracy_ovr:
    print(f"One-vs-One (OvO) performed better by: {(accuracy_ovo - accuracy_ovr):.4f}")
else:
    print("Both strategies achieved the same accuracy.")
````

````python
# Q18. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute theaverage accuracy=
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

# Standardize the features, which is crucial for SVMs
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize the SVM classifier
svm_classifier = SVC(kernel='linear', random_state=42)

# Initialize Stratified K-Fold Cross-Validation
# n_splits: number of folds (e.g., 5 or 10 are common)
# shuffle: whether to shuffle the data before splitting (important for randomness)
# random_state: for reproducibility
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# List to store accuracy for each fold
accuracies = []

print("Starting Stratified K-Fold Cross-Validation...")

# Perform cross-validation
for fold, (train_index, test_index) in enumerate(skf.split(X_scaled, y)):
    X_train, X_test = X_scaled[train_index], X_scaled[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Train the SVM model on the current fold's training data
    svm_classifier.fit(X_train, y_train)

    # Make predictions on the current fold's test data
    y_pred = svm_classifier.predict(X_test)

    # Calculate accuracy for the current fold
    fold_accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(fold_accuracy)
    print(f"  Fold {fold+1} Accuracy: {fold_accuracy:.4f}")

# Compute the average accuracy across all folds
average_accuracy = np.mean(accuracies)
std_accuracy = np.std(accuracies)

print(f"\nAverage Accuracy across all folds: {average_accuracy:.4f}")
print(f"Standard Deviation of Accuracy: {std_accuracy:.4f}")
````

````python
#Q21.Write a Python program to train a Naïve Bayes Classifier and evaluate its performance using Log Loss(Cross-Entropy Loss)=
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import log_loss
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Naïve Bayes models are generally not sensitive to feature scaling,
# but it's good practice for general machine learning pipelines.
# For GaussianNB, scaling can sometimes stabilize variance estimation,
# but often doesn't dramatically change results compared to SVMs or neural networks.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- Gaussian Naïve Bayes Classifier ---")
# Initialize the Gaussian Naïve Bayes classifier
gnb_classifier = GaussianNB()

# Train the classifier on the training data
gnb_classifier.fit(X_train_scaled, y_train)

# Predict probabilities for the test set
# Log Loss requires probability estimates, not just hard class predictions
y_pred_proba = gnb_classifier.predict_proba(X_test_scaled)

# Evaluate the model using Log Loss
# The true labels (y_test) and predicted probabilities (y_pred_proba) are needed.
loss = log_loss(y_test, y_pred_proba)
print(f"Log Loss of Gaussian Naïve Bayes Classifier: {loss:.4f}")

print("\nNote: A lower Log Loss value indicates better performance. It penalizes false classifications more severely when the predicted probability for the correct class is low.")
````

````python
#Q25. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUCscore=
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the Breast Cancer dataset
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features for potentially better performance, though not strictly required for GNB
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("--- Gaussian Naïve Bayes Classifier ---")
# Initialize the Gaussian Naïve Bayes classifier
gnb_classifier = GaussianNB()

# Train the classifier on the training data
gnb_classifier.fit(X_train_scaled, y_train)

# Predict probabilities for the positive class (class 1) on the test set
# ROC-AUC requires probability estimates, specifically for the positive class
y_pred_proba = gnb_classifier.predict_proba(X_test_scaled)[:, 1]

# Calculate the ROC-AUC score
roc_auc = roc_auc_score(y_test, y_pred_proba)
print(f"ROC-AUC Score of Gaussian Naïve Bayes Classifier: {roc_auc:.4f}")

# --- Optional: Visualize the ROC Curve ---
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

print("\nNote: A ROC-AUC score of 1.0 represents a perfect classifier, while 0.5 represents a random classifier. Higher scores indicate better discriminatory power.")
````

````python
#Q26. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, auc
from sklearn.preprocessing import StandardScaler

# Load the Breast Cancer dataset (a common dataset for binary classification)
cancer = datasets.load_breast_cancer()
X = cancer.data
y = cancer.target

# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features, which is crucial for SVMs
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Initialize and train the SVM classifier
# For precision_recall_curve, we need probability estimates or decision function scores.
# SVC with 'probability=True' enables probability estimates, but it adds computational cost.
# Alternatively, we can use `decision_function` which is faster and often sufficient.
svm_classifier = SVC(kernel='linear', random_state=42, probability=True)
# If probability=False, use decision_function:
# y_scores = svm_classifier.decision_function(X_test_scaled)

svm_classifier.fit(X_train_scaled, y_train)

# Get probability estimates for the positive class (class 1)
y_pred_proba = svm_classifier.predict_proba(X_test_scaled)[:, 1]

# Calculate precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_proba)

# Calculate the Area Under the Curve (AUC) for the Precision-Recall curve
pr_auc = auc(recall, precision)

# Visualize the Precision-Recall curve
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='darkorange', lw=2,
         label=f'Precision-Recall curve (AUC = {pr_auc:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve for SVM Classifier')
plt.legend(loc='lower left')
plt.grid(True)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.show()

print(f"Area Under the Precision-Recall Curve (PR-AUC): {pr_auc:.4f}")
print("\nNote: A higher PR-AUC score indicates better performance, especially in scenarios with class imbalance. It represents a good balance between precision and recall across different thresholds.")
````
